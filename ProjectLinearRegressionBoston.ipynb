```python
from sklearn.datasets import load_boston
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
boston = load_boston()
```

```python
x = boston.data[:,12]
y = boston.target

print(x.shape)
print(y.shape)
```

```python
data = pd.DataFrame(boston.data)
#print(data)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
mean_x = x.mean()
mean_y = y.mean()

m = len(x)
numer =0 
denom = 0
for i in range(m):
    numer +=(x[i] - mean_x)*(y[i] - mean_y)
    denom +=(x[i] - mean_x)**2
b1 =numer/denom
b0 =mean_y - (b1*mean_x)
print(b1,b0)
```

```python
x1 = np.linspace(0,35,10)
y1 = b0 + b1*x1

plt.plot(x1,y1,color='orange')
plt.scatter(x,y)
plt.show()
```
